```python
from mnist_loader import load_data
from network import Network
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [12, 12]
```

```python
training_data, validation_data, test_data = load_data()
```

```python
nw = Network([784, 30, 10])
```

```python
# train the network
nw.stoc_grad(training_data, 10, 10, 1.0, test_data)
```

```python
def predict(x):
    return np.argmax(nw.feed_forward(x))
```

```python
wrong_pred = [(x, y) for x, y in validation_data if predict(x) != y]
```

```python
def show(data, dim=8):
    fig, axes = plt.subplots(dim, dim, figsize=(10, 10))
    fig.tight_layout(pad=0.1, rect=[0, 0.03, 1, 0.92])
        
    for i, ax in enumerate(axes.flat):
        rand_index = np.random.randint(len(data))
        x, y = data[rand_index]
        ax.imshow(x.reshape(-1, 28), cmap=plt.get_cmap('gray'))
        ax.set_title(f'p: {predict(x)} | r: {y}')
        ax.set_axis_off()
    plt.show()
```

```python
show(wrong_pred, dim=6)
```
